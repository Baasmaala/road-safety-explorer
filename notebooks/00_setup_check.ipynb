```python
import pandas as pd

who = pd.read_excel(
    "data/raw/gsrrs23-indicators-for-participating-countries-or-territories.xlsx",
    sheet_name="Indicators"
)

deaths = pd.read_csv("data/raw/output.csv")

print("WHO indicators:", who.shape)
print("Deaths data:", deaths.shape)
```

```python
who_pal = who[who.apply(
    lambda r: r.astype(str).str.contains('Palestin|PSE', case=False, na=False).any(),
    axis=1
)]
deaths_pal = deaths[deaths['Entity'] == 'Palestine']

print("Palestine in WHO data:    ", len(who_pal), "row(s)")
print("Palestine in deaths data: ", len(deaths_pal), "row(s) (expect 30)")

assert len(who_pal) >= 1, "Palestine missing from WHO file!"
assert len(deaths_pal) == 30, "Palestine rows in deaths file unexpected!"
print(" Palestine verified in both datasets")
```
